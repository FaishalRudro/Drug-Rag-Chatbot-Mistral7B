```python
#MAINLY MAIN

import subprocess
import sys

def install_packages():
    print("üîß Installing required packages...")
    packages = [
        'openpyxl', 'pandas', 'sentence-transformers==3.3.1',
        'faiss-gpu', 'torch', 'rank_bm25', 'transformers',
        'accelerate', 'bitsandbytes', 'xlrd', 'tqdm'
    ]
    for pkg in packages:
        result = subprocess.run(
            [sys.executable, '-m', 'pip', 'install', '-q', pkg],
            capture_output=True
        )
    # faiss-gpu ‡¶®‡¶æ ‡¶•‡¶æ‡¶ï‡¶≤‡ßá faiss-cpu fallback
    try:
        import faiss
        _ = faiss.StandardGpuResources()
    except Exception:
        subprocess.run(
            [sys.executable, '-m', 'pip', 'install', '-q', 'faiss-cpu'],
            capture_output=True
        )
    print("‚úÖ All packages installed!\n")

install_packages()

# ============================================================================
# IMPORTS
# ============================================================================
import re
import gc
import time
import warnings
import multiprocessing as mp
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from typing import List, Dict, Optional, Tuple
from collections import defaultdict
from tqdm.auto import tqdm

import torch
import torch.cuda.amp as amp
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from sentence_transformers import SentenceTransformer, CrossEncoder
import faiss
from rank_bm25 import BM25Okapi

print("‚úÖ All imports successful!\n")

# ============================================================================
# CONFIGURATION
# ============================================================================
class Config:
    DATASET_PATH = "/kaggle/input/datasets/mdfaishalahmedrudroo/drugrecommendationdata/dataset_Final.xlsx"

    COLUMNS = {
        'drug_name':        'Drug Name',
        'generic_name':     'Generic Name',
        'drug_class':       'Drug Class',
        'uses':             'Uses',
        'warnings':         'Warnings',
        'use_instructions': 'Pre-Use Instructions',
        'dosage':           'Dosage',
        'side_effects':     'Side Effects'
    }

    # Embedding
    EMBEDDING_MODEL   = "BAAI/bge-large-en-v1.5"
    EMBEDDING_DIM     = 1024
    BGE_QUERY_PREFIX  = "Represent this sentence for searching relevant passages: "

    # Chunking
    # 300 words ‚âà 380 tokens ‚Äî BGE 512 limit-‡¶è‡¶∞ ‡¶Æ‡¶ß‡ßç‡¶Ø‡ßá safe
    CHUNK_MAX_WORDS   = 300
    CHUNK_OVERLAP     = 50

    # Encoding ‚Äî GPU optimized
    ENCODE_BATCH_SIZE  = 256     # V5: 128 ‚Üí 256 (GPU ~95% utilization)
    ENCODE_PRECISION   = 'fp16'  # half precision ‚Üí 2x GPU throughput
    NUM_WORKERS        = min(mp.cpu_count(), 8)  # parallel workers

    # Retrieval
    TOP_K_CHUNKS      = 20    # chunk-level candidates
    TOP_K_DRUGS       = 5     # drug-level final results
    HYBRID_ALPHA      = 0.65

    # LLM
    LLM_MODEL              = "mistralai/Mistral-7B-Instruct-v0.3"
    MAX_NEW_TOKENS         = 512

    # ‚îÄ‚îÄ Token budget (OOM prevention) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    # Mistral-7B context = 4096 tokens
    # System prompt + instruction + query ‚âà ~400 tokens
    # ‡¶§‡¶æ‡¶á context-‡¶è‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø ‡¶¨‡¶æ‡¶ï‡¶ø ‡¶•‡¶æ‡¶ï‡ßá ~2800 tokens
    # ‡¶è‡¶á budget-‡¶è‡¶∞ ‡¶Æ‡¶ß‡ßç‡¶Ø‡ßá chunks fit ‡¶ï‡¶∞‡¶æ ‡¶π‡¶¨‡ßá
    MAX_LLM_CONTEXT_TOKENS = 2800

    # ‚îÄ‚îÄ TEST MODE ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
    # None = ‡¶∏‡¶¨ drugs load ‡¶ï‡¶∞‡ßã (full dataset)
    # 100  = ‡¶∂‡ßÅ‡¶ß‡ßÅ ‡¶™‡ßç‡¶∞‡¶•‡¶Æ ‡ßß‡ß¶‡ß¶ drugs (test ‡¶ï‡¶∞‡¶æ‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø)
    MAX_DRUGS_TO_LOAD = None

config = Config()

# ============================================================================
# STEP 1: DATA LOADER + NOISE CLEANER
# ============================================================================
class DrugDataLoader:

    NULL_PATTERNS = re.compile(
        r'^(not\s+available|n/?a|nan|none|null|-+)$', re.IGNORECASE
    )
    BOILERPLATE_STARTS = [
        "get emergency medical help", "call 911",
        "seek emergency medical attention", "seek emergency",
        "follow all directions on your medicine label", "follow all directions",
    ]

    def __init__(self, path: str):
        self.path  = path
        self.df    = None
        self.drugs: List[Dict] = []

    @staticmethod
    def is_null_like(value) -> bool:
        if value is None or (isinstance(value, float) and np.isnan(value)):
            return True
        return bool(DrugDataLoader.NULL_PATTERNS.fullmatch(str(value).strip()))

    @staticmethod
    def clean_cell(value) -> str:
        if DrugDataLoader.is_null_like(value): return ""
        text = str(value).strip().strip('"\'')
        text = re.sub(r'[\r\n]+', ' ', text)
        text = re.sub(r'\s{2,}', ' ', text)
        text = re.sub(r'\bNot\s+Avail\w*\b', '', text, flags=re.IGNORECASE).strip()
        return text

    @staticmethod
    def clean_generic_name(value) -> str:
        if DrugDataLoader.is_null_like(value): return ""
        text = str(value).strip()
        text = re.sub(r'\[.*?\]', '', text)
        text = re.sub(
            r'\((?:oral|injection|topical|inhalation|solution|tablet|capsule)[^)]*\)',
            '', text, flags=re.IGNORECASE
        )
        text = re.sub(r'[\r\n].*', '', text)
        text = re.sub(r'Brand\s+names?:.*', '', text, flags=re.IGNORECASE)
        return re.sub(r'\s{2,}', ' ', text).strip()

    @staticmethod
    def clean_long_field(value) -> str:
        if DrugDataLoader.is_null_like(value): return ""
        text = str(value).strip()
        text = re.sub(r'\bNot\s+Avail\w*\b', '', text, flags=re.IGNORECASE)
        text = re.sub(r'([a-z])([A-Z])', r'\1 \2', text)
        text = re.sub(r'\.([A-Za-z])', r'. \1', text)
        text = re.sub(r'[\r\n]+', ' ', text)
        text = re.sub(r'\s{2,}', ' ', text).strip()

        sentences = re.split(r'(?<=[.!?])\s+', text)
        seen, deduped = set(), []
        for s in sentences:
            s = s.strip()
            key = s.lower()
            if len(s) > 8 and key not in seen:
                seen.add(key)
                deduped.append(s)

        filtered, bp_kept = [], False
        for s in deduped:
            is_bp = any(s.lower().startswith(b) for b in DrugDataLoader.BOILERPLATE_STARTS)
            if is_bp:
                if not bp_kept:
                    filtered.append(s)
                    bp_kept = True
            else:
                filtered.append(s)

        return " ".join(filtered).strip()

    @staticmethod
    def clean_dosage(value) -> str:
        if DrugDataLoader.is_null_like(value): return ""
        text = str(value).strip()
        first_line = text.split('\n')[0].strip()
        if DrugDataLoader.NULL_PATTERNS.fullmatch(first_line):
            parts = re.split(r'\n+', text, maxsplit=1)
            if len(parts) > 1:
                remainder = parts[1].strip()
                if any(remainder.lower().startswith(b) for b in DrugDataLoader.BOILERPLATE_STARTS):
                    return ""
                text = remainder
        return DrugDataLoader.clean_long_field(text)

    def load(self) -> List[Dict]:
        print("=" * 70)
        print("üìä STEP 1: LOADING & CLEANING DRUG DATASET")
        print("=" * 70)

        try:
            self.df = pd.read_excel(self.path, engine='openpyxl')
        except Exception:
            self.df = pd.read_excel(self.path, engine='xlrd')

        print(f"   Raw rows (before dedup) : {len(self.df):,}")

        # ‚îÄ‚îÄ DUPLICATE REMOVAL ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        # ‡¶∏‡¶¨ column ‡¶Æ‡¶ø‡¶≤‡¶ø‡¶Ø‡¶º‡ßá exactly identical rows ‚Üí ‡¶™‡ßç‡¶∞‡¶•‡¶Æ‡¶ü‡¶æ ‡¶∞‡ßá‡¶ñ‡ßá ‡¶¨‡¶æ‡¶ï‡¶ø ‡¶∏‡¶¨ drop
        # keep='first'  ‚Üí ‡¶™‡ßç‡¶∞‡¶•‡¶Æ occurrence ‡¶∞‡¶æ‡¶ñ‡ßã
        # inplace=True  ‚Üí df-‡¶è ‡¶∏‡¶∞‡¶æ‡¶∏‡¶∞‡¶ø apply ‡¶ï‡¶∞‡ßã
        rows_before = len(self.df)
        self.df.drop_duplicates(keep='first', inplace=True)
        self.df.reset_index(drop=True, inplace=True)
        rows_after   = len(self.df)
        duplicates_removed = rows_before - rows_after

        if duplicates_removed > 0:
            print(f"   ‚ôªÔ∏è  Duplicates removed   : {duplicates_removed:,} rows")
        else:
            print(f"   ‚úÖ No exact duplicates found")
        print(f"   Rows after dedup        : {rows_after:,}")
        # ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

        # ‚îÄ‚îÄ TEST MODE: ‡¶∂‡ßÅ‡¶ß‡ßÅ ‡¶™‡ßç‡¶∞‡¶•‡¶Æ N rows ‡¶®‡¶æ‡¶ì ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        if config.MAX_DRUGS_TO_LOAD is not None:
            self.df = self.df.head(config.MAX_DRUGS_TO_LOAD).copy()
            print(f"   ‚ö†Ô∏è  TEST MODE: ‡¶∂‡ßÅ‡¶ß‡ßÅ ‡¶™‡ßç‡¶∞‡¶•‡¶Æ {config.MAX_DRUGS_TO_LOAD} rows")
            print(f"   üí° Full dataset-‡¶è‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø config.MAX_DRUGS_TO_LOAD = None ‡¶ï‡¶∞‡ßÅ‡¶®\n")

        print(f"   Final rows : {len(self.df):,}")
        print(f"   Columns    : {list(self.df.columns)}\n")
        self._remap_columns()

        C = config.COLUMNS
        errors = 0

        # ‚îÄ‚îÄ Vectorized cleaning (iterrows() ‡¶è‡¶∞ ‡¶ö‡ßá‡¶Ø‡¶º‡ßá ~10x faster) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        print("   üßπ Vectorizing column cleaning...")

        # ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø column ‡¶Ü‡¶≤‡¶æ‡¶¶‡¶æ‡¶≠‡¶æ‡¶¨‡ßá pandas apply() ‡¶¶‡¶ø‡¶Ø‡¶º‡ßá clean ‡¶ï‡¶∞‡ßã
        col_map = {
            'drug_name':        (C.get('drug_name',''),        self.clean_cell),
            'generic_name':     (C.get('generic_name',''),     self.clean_generic_name),
            'drug_class':       (C.get('drug_class',''),       self.clean_cell),
            'uses':             (C.get('uses',''),             self.clean_long_field),
            'warnings':         (C.get('warnings',''),         self.clean_long_field),
            'dosage':           (C.get('dosage',''),           self.clean_dosage),
            'side_effects':     (C.get('side_effects',''),     self.clean_long_field),
            'use_instructions': (C.get('use_instructions',''), self.clean_long_field),
        }

        cleaned = {}
        for key, (col, fn) in tqdm(col_map.items(),
                                    desc="   üßπ Cleaning columns",
                                    unit="col", ncols=80, colour="green"):
            if col and col in self.df.columns:
                cleaned[key] = self.df[col].apply(fn)
            else:
                cleaned[key] = pd.Series([''] * len(self.df))

        # ‚îÄ‚îÄ Build records from cleaned series ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        print("\n   üì¶ Building drug records...")
        drug_name_series = cleaned['drug_name']
        valid_mask       = drug_name_series.str.len() > 0

        with tqdm(total=valid_mask.sum(), desc="   üì¶ Assembling records",
                  unit="drug", ncols=80, colour="cyan") as pbar:
            for idx in self.df.index[valid_mask]:
                try:
                    record = {
                        'drug_name':        cleaned['drug_name'][idx],
                        'generic_name':     cleaned['generic_name'][idx],
                        'drug_class':       cleaned['drug_class'][idx],
                        'uses':             cleaned['uses'][idx],
                        'warnings':         cleaned['warnings'][idx],
                        'dosage':           cleaned['dosage'][idx],
                        'side_effects':     cleaned['side_effects'][idx],
                        'use_instructions': cleaned['use_instructions'][idx],
                        'row_index':        idx,
                    }
                    self.drugs.append(record)
                except Exception:
                    errors += 1
                pbar.update(1)

        print(f"\n   ‚úÖ Loaded  : {len(self.drugs):,} valid drug records")
        if errors:
            print(f"   ‚ö†Ô∏è  Skipped : {errors:,} rows (errors)")
        print()
        return self.drugs

    def _remap_columns(self):
        df_lower   = {c.lower().strip(): c for c in self.df.columns}
        candidates = {
            'drug_name':        ['drug name', 'drugname', 'name'],
            'generic_name':     ['generic name', 'generic', 'genericname'],
            'drug_class':       ['drug class', 'class', 'category'],
            'uses':             ['uses', 'indications', 'use'],
            'warnings':         ['warnings', 'warning', 'precautions'],
            'use_instructions': ['pre-use instructions', 'pre use instructions',
                                 'use instructions', 'instructions', 'directions'],
            'dosage':           ['dosage', 'dose', 'dosing'],
            'side_effects':     ['side effects', 'adverse effects', 'side effect']
        }
        for key, tries in candidates.items():
            for t in tries:
                if t in df_lower:
                    config.COLUMNS[key] = df_lower[t]
                    break
        print(f"   Column map : {config.COLUMNS}\n")


# ============================================================================
# PARALLEL CHUNKING WORKER  (module-level ‚Äî pickle ‡¶ï‡¶∞‡¶æ ‡¶Ø‡¶æ‡¶Ø‡¶º)
# ============================================================================
def _chunk_one_drug_worker(args: Tuple) -> List[Dict]:
    """
    ProcessPoolExecutor-‡¶è ‡¶ö‡¶≤‡¶æ‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø module-level function‡•§
    drug_ref ‡¶¨‡¶æ‡¶¶‡ßá ‡¶∏‡¶¨ field ‡¶¶‡ßá‡¶Ø‡¶º ‚Äî drug_ref ‡¶™‡¶∞‡ßá attach ‡¶ï‡¶∞‡¶æ ‡¶π‡¶Ø‡¶º‡•§
    """
    drug, max_words, overlap = args
    name   = drug['drug_name']
    chunks = []

    FIELD_LABELS = {
        'uses':             'Drug uses and indications',
        'side_effects':     'Drug side effects and adverse reactions',
        'warnings':         'Drug warnings and contraindications',
        'dosage':           'Drug dosage and administration',
        'use_instructions': 'Drug pre-use instructions and precautions',
        'meta':             'Drug overview',
    }

    # META chunk
    meta_parts = [f"Drug name: {name}"]
    if drug['generic_name']:
        meta_parts.append(f"Generic name: {drug['generic_name']}")
    if drug['drug_class']:
        meta_parts.append(f"Drug class: {drug['drug_class']}")
    if drug['uses']:
        uses_preview = " ".join(drug['uses'].split()[:150])
        meta_parts.append(f"Primary uses: {uses_preview}")

    meta_text = " | ".join(meta_parts)
    chunks.append({
        'chunk_id': f"{name}::meta::0", 'drug_name': name,
        'field': 'meta', 'text': meta_text,
        'chunk_index': 0, 'total_chunks_for_field': 1,
    })

    # Long fields
    for field in ['uses', 'side_effects', 'warnings', 'dosage', 'use_instructions']:
        text = drug.get(field, '')
        if not text:
            continue

        if field == 'uses' and len(text.split()) <= 150:
            continue
        if field == 'uses':
            text = " ".join(text.split()[120:])

        # Split into chunks
        sentences   = re.split(r'(?<=[.!?])\s+', text.strip())
        sentences   = [s.strip() for s in sentences if len(s.strip()) > 5]
        field_chunks, current, current_len = [], [], 0

        for sent in sentences:
            words = sent.split()
            if current_len + len(words) > max_words and current:
                field_chunks.append(" ".join(current))
                overlap_words = " ".join(current).split()[-overlap:]
                current, current_len = list(overlap_words), len(overlap_words)
            current.extend(words)
            current_len += len(words)
        if current:
            field_chunks.append(" ".join(current))
        if not field_chunks:
            field_chunks = [text]

        label = FIELD_LABELS.get(field, field)
        total = len(field_chunks)
        for i, chunk_text in enumerate(field_chunks):
            chunks.append({
                'chunk_id': f"{name}::{field}::{i}", 'drug_name': name,
                'field': field, 'text': f"{label} of {name}: {chunk_text}",
                'chunk_index': i, 'total_chunks_for_field': total,
            })

    return chunks


# ============================================================================
# STEP 2: FIELD-WISE CHUNKER
# ============================================================================
class FieldChunker:
    """
    ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø drug field ‚Üí ‡¶Ü‡¶≤‡¶æ‡¶¶‡¶æ chunks‡•§
    ‡¶ï‡ßã‡¶®‡ßã information ‡¶¨‡¶æ‡¶¶ ‡¶Ø‡¶æ‡¶Ø‡¶º ‡¶®‡¶æ ‚Äî ‡¶™‡ßÅ‡¶∞‡ßã field text chunk ‡¶π‡¶Ø‡¶º‡•§

    chunk structure:
    {
        'chunk_id'   : "Abatacept::side_effects::2",
        'drug_name'  : "Abatacept",
        'field'      : "side_effects",
        'text'       : "Drug side effects of Abatacept: ...",
        'chunk_index': 2,
        'total_chunks_for_field': 3,   ‚Üê ‡¶è‡¶á field-‡¶è ‡¶Æ‡ßã‡¶ü ‡¶ï‡¶§‡¶ü‡¶ø chunk
        'drug_ref'   : <full drug dict>
    }
    """

    FIELD_LABELS = {
        'uses':             'Drug uses and indications',
        'side_effects':     'Drug side effects and adverse reactions',
        'warnings':         'Drug warnings and contraindications',
        'dosage':           'Drug dosage and administration',
        'use_instructions': 'Drug pre-use instructions and precautions',
        'meta':             'Drug overview',
    }

    def __init__(self):
        self.max_words = config.CHUNK_MAX_WORDS
        self.overlap   = config.CHUNK_OVERLAP

    def chunk_drugs(self, drugs: List[Dict]) -> List[Dict]:
        print("=" * 70)
        print("‚úÇÔ∏è  STEP 2: FIELD-WISE CHUNKING (PARALLEL)")
        print("=" * 70)
        print(f"   Max words/chunk : {self.max_words}")
        print(f"   Overlap words   : {self.overlap}")
        print(f"   Workers         : {config.NUM_WORKERS}\n")

        all_chunks   = []
        field_counts = defaultdict(int)

        chunker_args = [(drug, self.max_words, self.overlap) for drug in drugs]
        chunk_start  = time.time()

        with ProcessPoolExecutor(max_workers=config.NUM_WORKERS) as executor:
            futures = list(tqdm(
                executor.map(_chunk_one_drug_worker, chunker_args,
                             chunksize=max(1, len(drugs) // (config.NUM_WORKERS * 4))),
                total=len(drugs),
                desc="   ‚úÇÔ∏è  Chunking (parallel)",
                unit="drug", ncols=80, colour="cyan"
            ))

        print("\n   üîó Attaching drug references...")
        for drug, raw_chunks in tqdm(
            zip(drugs, futures),
            total=len(drugs),
            desc="   üîó Attaching refs",
            unit="drug", ncols=80, colour="blue"
        ):
            for chunk in raw_chunks:
                chunk['drug_ref'] = drug
                all_chunks.append(chunk)
                field_counts[chunk['field']] += 1

        chunk_elapsed = time.time() - chunk_start

        print(f"\n   {'‚îÄ'*50}")
        print(f"   Total chunks    : {len(all_chunks):,}")
        print(f"   Avg/drug        : {len(all_chunks)/max(len(drugs),1):.1f}")
        print(f"   Time            : {chunk_elapsed:.1f}s")
        print(f"   {'‚îÄ'*50}")
        for field, count in sorted(field_counts.items()):
            print(f"   ‚Ä¢ {field:<22}: {count:>8,} chunks")
        print(f"\n‚úÖ Chunking complete!\n")
        return all_chunks

    def _chunk_one_drug(self, drug: Dict) -> List[Dict]:
        chunks = []
        name   = drug['drug_name']

        meta_parts = [f"Drug name: {name}"]
        if drug['generic_name']:
            meta_parts.append(f"Generic name: {drug['generic_name']}")
        if drug['drug_class']:
            meta_parts.append(f"Drug class: {drug['drug_class']}")
        if drug['uses']:
            uses_preview = " ".join(drug['uses'].split()[:150])
            meta_parts.append(f"Primary uses: {uses_preview}")

        meta_text = " | ".join(meta_parts)
        chunks.append(self._make_chunk(name, 'meta', meta_text, 0, 1, drug))

        long_fields = ['uses', 'side_effects', 'warnings', 'dosage', 'use_instructions']

        for field in long_fields:
            text = drug.get(field, '')
            if not text:
                continue
            if field == 'uses' and len(text.split()) <= 150:
                continue
            if field == 'uses':
                text = " ".join(text.split()[120:])

            field_chunks = self._split_into_chunks(text)
            total        = len(field_chunks)
            label        = self.FIELD_LABELS.get(field, field)

            for i, chunk_text in enumerate(field_chunks):
                full_text = f"{label} of {name}: {chunk_text}"
                chunks.append(self._make_chunk(name, field, full_text, i, total, drug))

        return chunks

    def _split_into_chunks(self, text: str) -> List[str]:
        sentences = re.split(r'(?<=[.!?])\s+', text.strip())
        sentences = [s.strip() for s in sentences if len(s.strip()) > 5]

        if not sentences:
            return [text] if text else []

        chunks, current, current_len = [], [], 0

        for sent in sentences:
            words = sent.split()
            if current_len + len(words) > self.max_words and current:
                chunks.append(" ".join(current))
                overlap_words = " ".join(current).split()[-self.overlap:]
                current       = list(overlap_words)
                current_len   = len(overlap_words)
            current.extend(words)
            current_len += len(words)

        if current:
            chunks.append(" ".join(current))

        return chunks if chunks else [text]

    @staticmethod
    def _make_chunk(drug_name: str, field: str, text: str,
                    idx: int, total: int, drug_ref: Dict) -> Dict:
        return {
            'chunk_id':               f"{drug_name}::{field}::{idx}",
            'drug_name':              drug_name,
            'field':                  field,
            'text':                   text,
            'chunk_index':            idx,
            'total_chunks_for_field': total,
            'drug_ref':               drug_ref,
        }


# ============================================================================
# STEP 3: BGE EMBEDDER
# ============================================================================
class BGEEmbedder:

    def __init__(self):
        print("=" * 70)
        print("üßÆ STEP 3: LOADING BGE-LARGE EMBEDDING MODEL")
        print("=" * 70)

        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        print(f"   Device  : {self.device}")

        if self.device == 'cuda':
            torch.cuda.empty_cache()
            gc.collect()
            print(f"   GPU     : {torch.cuda.get_device_name(0)}")
            vram = torch.cuda.get_device_properties(0).total_memory / 1e9
            print(f"   VRAM    : {vram:.1f} GB")

        print(f"   Loading : {config.EMBEDDING_MODEL} ...")
        self.model = SentenceTransformer(config.EMBEDDING_MODEL, device=self.device)
        self.dim   = config.EMBEDDING_DIM
        print(f"   Dim     : {self.dim}")
        print(f"‚úÖ BGE Embedder ready!\n")

    def encode_documents(self, texts: List[str]) -> np.ndarray:
        """
        GPU-optimized encoding:
        ‚Ä¢ fp16 half precision ‚Üí GPU throughput 2x
        ‚Ä¢ batch_size=256 ‚Üí GPU ~95% utilization
        ‚Ä¢ torch.cuda.amp.autocast ‚Üí mixed precision ops
        ‚Ä¢ non_blocking CUDA transfers
        ‚Ä¢ GPU memory clear every 20 batches
        """
        batch_size = config.ENCODE_BATCH_SIZE
        total      = len(texts)
        use_fp16   = (self.device == 'cuda')

        print(f"   Encoding  : {total:,} chunks")
        print(f"   Batch size: {batch_size}")
        print(f"   Precision : {'fp16 (GPU accelerated)' if use_fp16 else 'fp32 (CPU)'}")
        print(f"   Batches   : {(total + batch_size - 1) // batch_size:,}\n")

        all_embs   = []
        start_time = time.time()

        if use_fp16:
            self.model.half()

        with tqdm(total=total, desc="   üî¢ Encoding",
                  unit="chunk", ncols=80, colour="yellow") as pbar:

            for i in range(0, total, batch_size):
                batch = texts[i : i + batch_size]

                if use_fp16:
                    with amp.autocast():
                        embs = self.model.encode(
                            batch,
                            normalize_embeddings=True,
                            show_progress_bar=False,
                            convert_to_numpy=True,
                            batch_size=batch_size,
                        )
                    embs = embs.astype(np.float32)
                else:
                    embs = self.model.encode(
                        batch,
                        normalize_embeddings=True,
                        show_progress_bar=False,
                        convert_to_numpy=True,
                        batch_size=batch_size,
                    )

                all_embs.append(embs)

                if self.device == 'cuda' and (i // batch_size) % 20 == 0:
                    torch.cuda.synchronize()
                    torch.cuda.empty_cache()

                pbar.update(len(batch))

                if i > 0 and (i // batch_size) % 10 == 0:
                    elapsed = time.time() - start_time
                    speed   = i / elapsed
                    eta_min = (total - i) / speed / 60 if speed > 0 else 0
                    if self.device == 'cuda':
                        mem_used = torch.cuda.memory_reserved() / 1e9
                        pbar.set_postfix({
                            'speed' : f"{speed:.0f}/s",
                            'ETA'   : f"{eta_min:.1f}m",
                            'VRAM'  : f"{mem_used:.1f}GB"
                        })
                    else:
                        pbar.set_postfix({
                            'speed': f"{speed:.0f}/s",
                            'ETA'  : f"{eta_min:.1f}m"
                        })

        if use_fp16:
            self.model.float()

        elapsed = time.time() - start_time
        result  = np.vstack(all_embs)

        print(f"\n   ‚úÖ Shape  : {result.shape}")
        print(f"   ‚è±Ô∏è  Time   : {elapsed/60:.1f} min")
        print(f"   üöÄ Speed  : {total/elapsed:.0f} chunks/sec\n")
        return result

    def encode_query(self, query: str) -> np.ndarray:
        prefixed = config.BGE_QUERY_PREFIX + query
        return self.model.encode(
            [prefixed],
            normalize_embeddings=True,
            convert_to_numpy=True
        )[0]


# ============================================================================
# STEP 4: FAISS INDEX (GPU-accelerated ‡¶Ø‡¶¶‡¶ø faiss-gpu ‡¶•‡¶æ‡¶ï‡ßá)
# ============================================================================
def build_faiss_index(embeddings: np.ndarray) -> faiss.Index:
    print("=" * 70)
    print("üóÑÔ∏è  STEP 4a: BUILDING FAISS INDEX")
    print("=" * 70)

    total = embeddings.shape[0]
    dim   = embeddings.shape[1]
    data  = embeddings.astype('float32')

    use_gpu_faiss = False
    try:
        res   = faiss.StandardGpuResources()
        cfg   = faiss.GpuIndexFlatConfig()
        cfg.useFloat16 = True
        index = faiss.GpuIndexFlatIP(res, dim, cfg)
        use_gpu_faiss = True
        print(f"   Mode    : GPU FAISS (fp16) üöÄ")
    except Exception:
        index = faiss.IndexFlatIP(dim)
        print(f"   Mode    : CPU FAISS")

    with tqdm(total=total, desc="   üìå Indexing vectors",
              unit="vec", ncols=80, colour="magenta") as pbar:
        chunk_size = 50_000
        for i in range(0, total, chunk_size):
            batch = data[i : i + chunk_size]
            index.add(batch)
            pbar.update(len(batch))

    print(f"\n   ‚úÖ FAISS : {index.ntotal:,} vectors indexed")
    print(f"   GPU FAISS: {'Yes ‚úÖ' if use_gpu_faiss else 'No (CPU fallback)'}\n")
    return index


# ============================================================================
# STEP 5: BM25 INDEX (parallel tokenization)
# ============================================================================
def build_bm25_index(chunks: List[Dict]) -> BM25Okapi:
    print("=" * 70)
    print("üìù STEP 4b: BUILDING BM25 INDEX (PARALLEL)")
    print("=" * 70)
    print(f"   Workers : {config.NUM_WORKERS}\n")

    texts = [c['text'] for c in chunks]

    def tokenize_batch(batch_texts):
        return [t.lower().split() for t in batch_texts]

    batch_size  = max(1, len(texts) // (config.NUM_WORKERS * 4))
    batches     = [texts[i:i+batch_size] for i in range(0, len(texts), batch_size)]
    tokenized   = []

    with ThreadPoolExecutor(max_workers=config.NUM_WORKERS) as executor:
        results = list(tqdm(
            executor.map(tokenize_batch, batches),
            total=len(batches),
            desc="   üìù Tokenizing (parallel)",
            unit="batch", ncols=80, colour="blue"
        ))

    for batch_result in results:
        tokenized.extend(batch_result)

    print(f"\n   Building BM25 index...")
    bm25 = BM25Okapi(tokenized)
    print(f"   ‚úÖ BM25 : {len(tokenized):,} documents\n")
    return bm25


# ============================================================================
# STEP 6: CHUNK RETRIEVER
# ============================================================================
class ChunkRetriever:

    PRIORITY_FIELDS = {
        'side_effects':  {'side_effects', 'meta'},
        'dosage_query':  {'dosage', 'meta'},
        'warning_query': {'warnings', 'use_instructions', 'meta'},
        'symptom_query': {'uses', 'meta'},
        'drug_lookup':   {'meta', 'uses', 'side_effects',
                          'warnings', 'dosage', 'use_instructions'},
        'general':       {'meta', 'uses'},
    }

    def __init__(self, chunks: List[Dict],
                 faiss_index: faiss.Index,
                 bm25: BM25Okapi):
        print("=" * 70)
        print("üîç STEP 5: INITIALIZING RETRIEVER")
        print("=" * 70)

        self.chunks      = chunks
        self.faiss_index = faiss_index
        self.bm25        = bm25

        print("   Loading cross-encoder reranker...")
        self.reranker = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')

        self.drug_to_chunks: Dict[str, List[int]] = defaultdict(list)
        for idx, chunk in enumerate(chunks):
            self.drug_to_chunks[chunk['drug_name']].append(idx)

        print(f"   Drugs     : {len(self.drug_to_chunks):,}")
        print(f"   Chunks    : {len(chunks):,}")
        print(f"‚úÖ Retriever ready!\n")

    def search(self, query: str, query_emb: np.ndarray,
               intent: str = 'general') -> List[Dict]:
        exact_names     = {n for n in self.drug_to_chunks if n.lower() in query.lower()}
        priority_fields = self.PRIORITY_FIELDS.get(intent, {'meta', 'uses'})

        n_search = min(config.TOP_K_CHUNKS * 6, len(self.chunks))
        q        = query_emb.reshape(1, -1).astype('float32')
        sem_scores, sem_idxs = self.faiss_index.search(q, n_search)
        sem_dict = {int(i): float(s) for i, s in zip(sem_idxs[0], sem_scores[0])}

        bm25_scores = self.bm25.get_scores(query.lower().split())
        max_bm25    = bm25_scores.max() if bm25_scores.max() > 0 else 1.0

        chunk_scores = {}
        for idx in range(len(self.chunks)):
            sem   = sem_dict.get(idx, 0.0)
            bm25  = float(bm25_scores[idx]) / max_bm25
            score = config.HYBRID_ALPHA * sem + (1 - config.HYBRID_ALPHA) * bm25

            if self.chunks[idx]['field'] in priority_fields:
                score *= 1.3
            if self.chunks[idx]['drug_name'] in exact_names:
                score += 0.4

            chunk_scores[idx] = score

        drug_best: Dict[str, Dict] = {}
        for chunk_idx, score in sorted(chunk_scores.items(),
                                        key=lambda x: x[1], reverse=True):
            if len(drug_best) > config.TOP_K_CHUNKS * 3:
                break

            chunk     = self.chunks[chunk_idx]
            drug_name = chunk['drug_name']

            if drug_name not in drug_best:
                drug_best[drug_name] = {
                    'drug_ref':      chunk['drug_ref'],
                    'matched_chunks': [],
                    'best_score':    score,
                }

            if len(drug_best[drug_name]['matched_chunks']) < 6:
                drug_best[drug_name]['matched_chunks'].append({
                    'field': chunk['field'],
                    'text':  chunk['text'],
                    'score': score,
                })

        sorted_drugs = sorted(
            drug_best.values(),
            key=lambda x: x['best_score'],
            reverse=True
        )[:config.TOP_K_CHUNKS]

        return self._rerank(query, sorted_drugs, config.TOP_K_DRUGS)

    def _rerank(self, query: str, drug_results: List[Dict],
                top_k: int) -> List[Dict]:
        if len(drug_results) <= top_k:
            return drug_results

        pairs = []
        for dr in drug_results:
            best_text = dr['matched_chunks'][0]['text'] if dr['matched_chunks'] else ""
            pairs.append([query, best_text[:700]])

        scores = self.reranker.predict(pairs)
        for i, dr in enumerate(drug_results):
            dr['rerank_score'] = float(scores[i])

        return sorted(drug_results,
                       key=lambda x: x['rerank_score'], reverse=True)[:top_k]


# ============================================================================
# STEP 7: TOKEN-BUDGET AWARE CONTEXT BUILDER
# ============================================================================
class SmartContextBuilder:

    FIELD_PRIORITY = {
        'side_effects':  ['side_effects', 'warnings', 'meta'],
        'dosage_query':  ['dosage', 'use_instructions', 'meta'],
        'warning_query': ['warnings', 'use_instructions', 'uses', 'meta'],
        'symptom_query': ['uses', 'meta'],
        'drug_lookup':   ['meta', 'uses', 'dosage', 'side_effects',
                          'warnings', 'use_instructions'],
        'general':       ['meta', 'uses', 'side_effects'],
    }

    @classmethod
    def build(cls, drug_results: List[Dict], intent: str,
              tokenizer) -> Tuple[str, int]:
        field_priority = cls.FIELD_PRIORITY.get(intent, ['meta', 'uses'])
        budget         = config.MAX_LLM_CONTEXT_TOKENS
        used_tokens    = 0
        context_parts  = []

        for dr in drug_results:
            drug = dr['drug_ref']
            name = drug['drug_name']

            header_lines = [f"‚îÅ‚îÅ‚îÅ {name} ‚îÅ‚îÅ‚îÅ"]
            if drug.get('generic_name'):
                header_lines.append(f"Generic: {drug['generic_name']}")
            if drug.get('drug_class'):
                header_lines.append(f"Class: {drug['drug_class']}")
            header_text   = "\n".join(header_lines)
            header_tokens = cls._count_tokens(header_text, tokenizer)

            if used_tokens + header_tokens > budget:
                break

            context_parts.append(header_text)
            used_tokens += header_tokens

            sorted_chunks = cls._sort_chunks_by_priority(
                dr['matched_chunks'], field_priority
            )

            for chunk_info in sorted_chunks:
                chunk_text   = chunk_info['text']
                chunk_tokens = cls._count_tokens(chunk_text, tokenizer)

                if used_tokens + chunk_tokens > budget:
                    continue

                context_parts.append(chunk_text)
                used_tokens += chunk_tokens

            context_parts.append("")

        return "\n".join(context_parts), used_tokens

    @staticmethod
    def _sort_chunks_by_priority(chunks: List[Dict],
                                  field_priority: List[str]) -> List[Dict]:
        def priority_key(chunk):
            field = chunk['field']
            try:
                return field_priority.index(field)
            except ValueError:
                return len(field_priority)
        return sorted(chunks, key=priority_key)

    @staticmethod
    def _count_tokens(text: str, tokenizer) -> int:
        if tokenizer is not None:
            try:
                return len(tokenizer.encode(text, add_special_tokens=False))
            except Exception:
                pass
        return int(len(text.split()) * 1.3)


# ============================================================================
# STEP 8: LLM
# ============================================================================
class LLMModel:

    def __init__(self):
        print("=" * 70)
        print("ü§ñ STEP 6: LOADING LLM")
        print("=" * 70)
        print(f"   Model : {config.LLM_MODEL}")
        print(f"   Quant : 4-bit NF4\n")

        torch.cuda.empty_cache()
        gc.collect()

        bnb_cfg = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_quant_type="nf4",
            bnb_4bit_compute_dtype=torch.float16,
        )
        self.tokenizer = AutoTokenizer.from_pretrained(config.LLM_MODEL)
        self.llm = AutoModelForCausalLM.from_pretrained(
            config.LLM_MODEL,
            quantization_config=bnb_cfg,
            device_map="auto",
            torch_dtype=torch.float16,
            max_memory={0: "10GB"}
        )
        self.llm.eval()
        print("‚úÖ LLM ready!\n")

    def generate(self, prompt: str) -> str:
        inputs = self.tokenizer(
            prompt,
            return_tensors="pt",
            truncation=True,
            max_length=4096
        ).to(self.llm.device)

        input_len = inputs['input_ids'].shape[1]

        with torch.no_grad():
            out = self.llm.generate(
                **inputs,
                max_new_tokens=config.MAX_NEW_TOKENS,
                temperature=0.1,
                top_p=0.9,
                do_sample=True,
                pad_token_id=self.tokenizer.eos_token_id,
                repetition_penalty=1.2
            )

        new_tokens = out[0][input_len:]
        return self.tokenizer.decode(new_tokens, skip_special_tokens=True).strip()


# ============================================================================
# STEP 9: INTENT CLASSIFIER
# ============================================================================
class IntentClassifier:
    PATTERNS = {
        'side_effects':  [r'\bside\s+effects?\b', r'\badverse\s+effects?\b',
                          r'\bunwanted\s+effects?\b'],
        'dosage_query':  [r'\bdosage\b', r'\bdose\b', r'\bhow\s+much\b',
                          r'\bhow\s+often\b', r'\bwhen\s+to\s+take\b'],
        'warning_query': [r'\bwarning\b', r'\bcontraindic\b', r'\bavoid\b.*\btaking\b',
                          r'\bsafe\b.*\bpregnant\b', r'\bsafe\b.*\bkidney\b',
                          r'\bsafe\b.*\bliver\b', r'\binteract(?:ion)?\b'],
        'symptom_query': [r'\btreat\b', r'\bcure\b', r'\bmedic(?:ine|ation)?\s+for\b',
                          r'\bsuffering\s+from\b', r'\bdiagnosed\s+with\b',
                          r'\bdrug\s+for\b', r'\bI\s+have\b', r'\bI\s+feel\b',
                          r'\brecommend\b', r'\bsuggest\b'],
        'drug_lookup':   [r'\bwhat\s+is\b', r'\btell\s+me\s+about\b',
                          r'\binfo\w*\s+(?:on|about)\b',
                          r'\bdetails?\s+(?:of|about|on)\b'],
    }

    @classmethod
    def classify(cls, query: str) -> str:
        q = query.lower()
        for intent, patterns in cls.PATTERNS.items():
            for p in patterns:
                if re.search(p, q):
                    return intent
        if re.search(r'\b[A-Z][a-z]+(?:\s+[A-Z][a-z]+)?\b', query):
            return 'drug_lookup'
        return 'general'


# ============================================================================
# CONVERSATION MEMORY (LSTM-inspired: Short-Term + Long-Term)
# ============================================================================
class ConversationMemory:
    """
    LSTM-‡¶è‡¶∞ ‡¶Æ‡¶§‡ßã ‡¶¶‡ßÅ‡¶ü‡ßã memory gate:

    SHORT-TERM MEMORY (recent turns):
    ‚îÄ ‡¶∂‡ßá‡¶∑ N turns ‡¶π‡ßÅ‡¶¨‡¶π‡ßÅ ‡¶∞‡¶æ‡¶ñ‡¶æ ‡¶π‡¶Ø‡¶º
    ‚îÄ LLM-‡¶è ‡¶∏‡¶∞‡¶æ‡¶∏‡¶∞‡¶ø conversation history ‡¶π‡¶ø‡¶∏‡ßá‡¶¨‡ßá ‡¶Ø‡¶æ‡¶Ø‡¶º
    ‚îÄ Recent context ‡¶ß‡¶∞‡ßá ‡¶∞‡¶æ‡¶ñ‡ßá ("its warnings?" ‚Üí ‡¶ï‡ßã‡¶® drug ‡¶¨‡ßã‡¶ù‡¶æ‡¶Ø‡¶º)

    LONG-TERM MEMORY (compressed summary):
    ‚îÄ ‡¶™‡ßÅ‡¶∞‡¶æ‡¶®‡ßã turns compress ‡¶ï‡¶∞‡ßá summary ‡¶∞‡¶æ‡¶ñ‡¶æ ‡¶π‡¶Ø‡¶º
    ‚îÄ ‡¶ï‡ßã‡¶® drugs ‡¶ú‡¶ø‡¶ú‡ßç‡¶û‡ßá‡¶∏ ‡¶ï‡¶∞‡¶æ ‡¶π‡¶Ø‡¶º‡ßá‡¶õ‡ßá, ‡¶ï‡ßã‡¶® topics cover ‡¶π‡¶Ø‡¶º‡ßá‡¶õ‡ßá
    ‚îÄ Session ‡¶∂‡ßÅ‡¶∞‡ßÅ‡¶∞ context ‡¶π‡¶æ‡¶∞‡¶æ‡¶Ø‡¶º ‡¶®‡¶æ
    ‚îÄ Token budget ‡¶¨‡¶æ‡¶Å‡¶ö‡¶æ‡¶Ø‡¶º (‡¶™‡ßÅ‡¶∞‡ßã history ‡¶¶‡¶ø‡¶§‡ßá ‡¶π‡¶Ø‡¶º ‡¶®‡¶æ)

    FORGET GATE (LSTM-inspired):
    ‚îÄ ‡¶Ö‡¶™‡ßç‡¶∞‡¶æ‡¶∏‡¶ô‡ßç‡¶ó‡¶ø‡¶ï/‡¶™‡ßÅ‡¶∞‡¶æ‡¶®‡ßã ‡¶§‡¶•‡ßç‡¶Ø long-term ‡¶•‡ßá‡¶ï‡ßá ‡¶∏‡¶∞‡¶æ‡¶®‡ßã ‡¶π‡¶Ø‡¶º
    ‚îÄ ‡¶®‡¶§‡ßÅ‡¶® drug topic ‡¶è‡¶≤‡ßá ‡¶™‡ßÅ‡¶∞‡¶æ‡¶®‡ßã drug summary compress ‡¶π‡¶Ø‡¶º
    """

    def __init__(self,
                 short_term_size: int = 4,
                 long_term_size:  int = 8):
        self.short_term_size = short_term_size
        self.long_term_size  = long_term_size

        self.short_term: List[Dict] = []
        self.long_term: Dict = {
            'drugs_discussed':  [],
            'topics_covered':   [],
            'last_drug':        None,
            'session_summary':  "",
        }
        self.full_history: List[Dict] = []

    def add_turn(self, user_msg: str, assistant_msg: str,
                 drugs_retrieved: List[Dict], intent: str):
        turn = {
            'user':      user_msg,
            'assistant': assistant_msg,
            'drugs':     [d['drug_ref']['drug_name'] for d in drugs_retrieved],
            'intent':    intent,
        }
        self.full_history.append(turn)
        self.short_term.append(turn)

        while len(self.short_term) > self.short_term_size:
            old_turn = self.short_term.pop(0)
            self._compress_to_long_term(old_turn)

        if turn['drugs']:
            for drug in turn['drugs']:
                if drug not in self.long_term['drugs_discussed']:
                    self.long_term['drugs_discussed'].append(drug)
            self.long_term['last_drug'] = turn['drugs'][0]

        if intent not in self.long_term['topics_covered']:
            self.long_term['topics_covered'].append(intent)

        if len(self.long_term['drugs_discussed']) > self.long_term_size:
            self.long_term['drugs_discussed'] = \
                self.long_term['drugs_discussed'][-self.long_term_size:]

    def _compress_to_long_term(self, turn: Dict):
        drug_str = ", ".join(turn['drugs']) if turn['drugs'] else "general query"
        summary_piece = f"[{turn['intent']}] about {drug_str}"
        if self.long_term['session_summary']:
            self.long_term['session_summary'] += f"; {summary_piece}"
        else:
            self.long_term['session_summary'] = summary_piece

    def get_context_for_llm(self) -> Tuple[str, str]:
        short_lines = []
        for turn in self.short_term:
            short_lines.append(f"User: {turn['user']}")
            short_lines.append(f"Assistant: {turn['assistant']}")
            short_lines.append("")
        short_term_str = "\n".join(short_lines).strip()

        long_parts = []
        if self.long_term['last_drug']:
            long_parts.append(
                f"Most recently discussed drug: {self.long_term['last_drug']}"
            )
        if self.long_term['drugs_discussed']:
            drugs_list = ", ".join(self.long_term['drugs_discussed'][-5:])
            long_parts.append(f"Drugs discussed in this session: {drugs_list}")
        if self.long_term['session_summary']:
            long_parts.append(
                f"Earlier conversation summary: {self.long_term['session_summary']}"
            )
        long_term_str = "\n".join(long_parts)
        return short_term_str, long_term_str

    def resolve_pronoun(self, query: str) -> str:
        pronouns = [
            r'\bits\b', r'\bthis drug\b', r'\bthis medication\b',
            r'\bthe drug\b', r'\bthe medication\b', r'\bit\b(?!\s+is)',
            r'\bthis\b', r'\bthat drug\b', r'\bsame drug\b',
        ]
        last_drug = self.long_term['last_drug']
        if not last_drug:
            return query

        q_lower = query.lower()
        for pattern in pronouns:
            if re.search(pattern, q_lower):
                resolved = re.sub(pattern, last_drug, query, flags=re.IGNORECASE)
                print(f"   üîó Pronoun resolved: '{query}' ‚Üí '{resolved}'")
                return resolved
        return query

    def clear(self):
        self.short_term  = []
        self.full_history = []
        self.long_term = {
            'drugs_discussed': [],
            'topics_covered':  [],
            'last_drug':       None,
            'session_summary': "",
        }


# ============================================================================
# STEP 10: PROMPT BUILDER
# ============================================================================
class PromptBuilder:

    SYSTEM = (
        "You are a helpful medical information assistant. "
        "Provide accurate drug information based ONLY on the provided context. "
        "Always remind users to consult a doctor or pharmacist before taking any medication. "
        "Never invent information not in the context. "
        "If information is unavailable, say 'I don't have that information.'"
    )

    INSTRUCTIONS = {
        'symptom_query':  ("Suggest relevant medications based on symptoms from context. "
                           "State drug name, what it treats, and its class. "
                           "Add: 'Please consult a doctor before taking any medication.'"),
        'side_effects':   ("List ALL side effects from context clearly. "
                           "Highlight serious side effects requiring emergency care."),
        'dosage_query':   ("Provide complete dosage information from context. "
                           "Emphasize dosage must be confirmed with a doctor."),
        'warning_query':  ("List ALL warnings, contraindications, and precautions. "
                           "Recommend consulting a healthcare provider."),
        'drug_lookup':    "Answer using the context. Be complete, clear, and accurate.",
        'general':        "Answer using the context. Be clear and accurate.",
    }

    @classmethod
    def build(cls, query: str, context: str, intent: str,
              memory: 'ConversationMemory') -> str:
        instruction = cls.INSTRUCTIONS.get(intent, cls.INSTRUCTIONS['general'])
        short_term_str, long_term_str = memory.get_context_for_llm()

        long_term_block = ""
        if long_term_str:
            long_term_block = (
                f"SESSION MEMORY (long-term context):\n{long_term_str}\n\n"
            )

        short_term_block = ""
        if short_term_str:
            short_term_block = (
                f"RECENT CONVERSATION (short-term memory):\n{short_term_str}\n\n"
            )

        return (
            f"<s>[INST] {cls.SYSTEM}\n\n"
            f"INSTRUCTION: {instruction}\n\n"
            f"{long_term_block}"
            f"{short_term_block}"
            f"DRUG INFORMATION CONTEXT:\n{context}\n\n"
            f"USER QUESTION: {query}\n\n"
            f"Provide a helpful, accurate, and complete response. [/INST]"
        )


# ============================================================================
# STEP 11: MAIN RAG SYSTEM
# ============================================================================
class DrugRAGSystem:

    def __init__(self, llm: LLMModel, retriever: ChunkRetriever,
                 embedder: BGEEmbedder, drugs: List[Dict]):
        self.llm       = llm
        self.retriever = retriever
        self.embedder  = embedder
        self.drugs     = drugs
        self.memory    = ConversationMemory(
            short_term_size=4,
            long_term_size=8
        )

        print("=" * 70)
        print("‚úÖ DRUG RAG SYSTEM V5.0 READY!")
        if config.MAX_DRUGS_TO_LOAD:
            print(f"   ‚ö†Ô∏è  TEST MODE : {config.MAX_DRUGS_TO_LOAD} drugs only")
        print(f"   Drugs     : {len(drugs):,}")
        print(f"   Embedding : {config.EMBEDDING_MODEL}")
        print(f"   LLM       : {config.LLM_MODEL}")
        print(f"   Memory    : short-term=4 turns | long-term=8 drugs")
        print(f"   Token budget (context): {config.MAX_LLM_CONTEXT_TOKENS}")
        print("=" * 70 + "\n")

    def ask(self, query: str, show_details: bool = False) -> Dict:
        print(f"\n{'='*70}")
        print(f"‚ùì Query  : {query}")
        print(f"{'='*70}")

        resolved_query = self.memory.resolve_pronoun(query)
        intent    = IntentClassifier.classify(resolved_query)
        print(f"üéØ Intent : {intent.upper()}")

        query_emb = self.embedder.encode_query(resolved_query)

        print("üîç Retrieving relevant chunks...")
        results = self.retriever.search(resolved_query, query_emb, intent)
        print(f"   Found {len(results)} relevant drugs")

        if not results:
            answer = ("I couldn't find relevant drug information. "
                      "Please consult a pharmacist or doctor.")
            self.memory.add_turn(query, answer, [], intent)
            return {'question': query, 'answer': answer, 'drugs': []}

        print("üìù Building context (token-budget aware)...")
        context, tokens_used = SmartContextBuilder.build(
            results, intent, self.llm.tokenizer
        )
        print(f"   Context tokens used : {tokens_used} / {config.MAX_LLM_CONTEXT_TOKENS}")

        prompt = PromptBuilder.build(resolved_query, context, intent, self.memory)
        print("üí≠ Generating response...")
        answer = self.llm.generate(prompt)

        if intent == 'symptom_query' and 'consult' not in answer.lower():
            answer += ("\n\n‚ö†Ô∏è Important: Please consult a licensed doctor or "
                       "pharmacist before starting any medication.")

        print(f"\n‚úÖ Answer:\n{answer}")

        if show_details:
            print(f"\n{'‚îÄ'*50}")
            print("üìä Retrieved drugs:")
            for i, r in enumerate(results, 1):
                name   = r['drug_ref']['drug_name']
                score  = r.get('rerank_score', r.get('best_score', 0))
                fields = [c['field'] for c in r['matched_chunks']]
                print(f"   {i}. {name}")
                print(f"      Score  : {score:.4f}")
                print(f"      Fields : {fields}")
            print(f"\nüß† Memory state:")
            print(f"   Short-term turns : {len(self.memory.short_term)}")
            print(f"   Last drug        : {self.memory.long_term['last_drug']}")
            print(f"   Drugs discussed  : {self.memory.long_term['drugs_discussed']}")

        self.memory.add_turn(query, answer, results, intent)
        print(f"{'='*70}\n")
        return {'question': query, 'answer': answer, 'drugs': results}

    def chat(self):
        print("\n" + "=" * 70)
        print("üí¨ DRUG INFO CHATBOT V5")
        print("   'quit' ‚Üí exit  |  'clear' ‚Üí reset memory")
        print("=" * 70 + "\n")

        while True:
            try:
                user_input = input("You: ").strip()
            except (EOFError, KeyboardInterrupt):
                print("\nGoodbye!"); break

            if not user_input: continue
            if user_input.lower() in ('quit', 'exit', 'q'):
                print("Goodbye!"); break
            if user_input.lower() == 'clear':
                self.memory.clear()
                print("‚úÖ Memory cleared.\n"); continue

            self.ask(user_input)


# ============================================================================
# BUILDER
# ============================================================================
def build_drug_rag(dataset_path: str, hf_token: str = None) -> DrugRAGSystem:

    total_start = time.time()

    print("\n" + "=" * 70)
    print("üöÄ DRUG RAG SYSTEM V5.0 ‚Äî BUILD START")
    if config.MAX_DRUGS_TO_LOAD:
        print(f"   ‚ö†Ô∏è  TEST MODE: ‡¶™‡ßç‡¶∞‡¶•‡¶Æ {config.MAX_DRUGS_TO_LOAD} drugs")
        print(f"   üí° Full run: config.MAX_DRUGS_TO_LOAD = None")
    print("   GPU-optimized | Parallel chunking | fp16 encoding | GPU FAISS")
    print("=" * 70 + "\n")

    if hf_token:
        from huggingface_hub import login
        print("üîê Logging into HuggingFace...")
        login(hf_token)
        print("‚úÖ Login successful!\n")

    # Step 1 ‚Äî Load & clean (duplicate rows ‡¶è‡¶ñ‡¶æ‡¶®‡ßá‡¶á remove ‡¶π‡¶Ø‡¶º)
    loader = DrugDataLoader(dataset_path)
    drugs  = loader.load()
    if not drugs:
        raise RuntimeError("No drug records loaded!")

    # Step 2 ‚Äî Chunk
    chunker = FieldChunker()
    chunks  = chunker.chunk_drugs(drugs)

    # Step 3 ‚Äî Embed
    embedder    = BGEEmbedder()
    chunk_texts = [c['text'] for c in chunks]
    embeddings  = embedder.encode_documents(chunk_texts)

    # Step 4a ‚Äî FAISS
    faiss_index = build_faiss_index(embeddings)

    # Step 4b ‚Äî BM25
    bm25 = build_bm25_index(chunks)

    # Step 5 ‚Äî Retriever
    retriever = ChunkRetriever(chunks, faiss_index, bm25)

    # Step 6 ‚Äî LLM
    llm = LLMModel()

    # Step 7 ‚Äî Assemble
    rag = DrugRAGSystem(llm, retriever, embedder, drugs)

    elapsed = time.time() - total_start
    print(f"üèÅ Total build time: {elapsed/60:.1f} minutes\n")
    return rag


# ============================================================================
# MAIN
# ============================================================================
if __name__ == "__main__":

    DATASET_PATH = "/kaggle/input/datasets/mdfaishalahmedrudroo/drugrecommendationdata/dataset_Final.xlsx"
    HF_TOKEN     = "hf_token"

    rag = build_drug_rag(DATASET_PATH, HF_TOKEN)

    print("\n" + "=" * 70)
    print("üß™ TEST QUERIES")
    print("=" * 70)

    test_queries = [
        "Tell me about Abilify",
        "What is Abatacept used for?",
        "What are the side effects of Abatacept?",
        "What is the dosage for Abreva?",
        "What are the pre-use instructions for Abatacept?",
        "I have severe pain and inflammation, what helps?",
        "What drugs are used for depression?",
        "Is Abstral safe for elderly patients?",
        "What about its warnings?",
    ]

    for q in test_queries:
        rag.ask(q, show_details=True)
        print()

    # rag.chat()   # interactive mode

    print("\n" + "=" * 70)
    print("‚úÖ DRUG RAG SYSTEM V5.0 COMPLETE!")
    print("=" * 70)
```
